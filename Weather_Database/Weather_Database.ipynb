```python
# Import Dependencies
import pandas as pd
import numpy as np
import requests
from citipy import citipy
import time 

# Import API keys
from config import weather_api_key
```

```python
# Create a set of random latitude and longitude combinations.
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lats_lngs = zip(lats, lngs)
coordinates = list(lats_lngs) 
len(coordinates)
```

```python
# Get the nearest cities to each coordinate with citipy
cities = list(set([citipy.nearest_city(coor[0],coor[1]).city_name for coor in coordinates]))
cities[:5]
```

```python
# init lists
countries, longs, lats, max_temps, humidities, cloudiness_per, wind_speeds, weather_desc = ([] for i in range(8))
# countries = []
# longs = []
# lats = []
# max_temps = []
# humidities = []
# cloudiness_per = []
# wind_speeds = []
# weather_desc = []


base_url = "http://api.openweathermap.org/data/2.5/weather"

for city in cities:

    # Build query URL
    query_url = f"{base_url}?appid={weather_api_key}&q={city}&units=imperial"
    
    response = requests.get(query_url)

    if (response.status_code == 200):
        data = response.json()
        
        # data parsing
        try:
            # print(data["sys"]["country"])
            countries.append(data["sys"]["country"])
            longs.append(data["coord"]["lon"])
            lats.append(data["coord"]["lat"])
            max_temps.append(data["main"]["temp_max"])
            humidities.append(data["main"]["humidity"])
            cloudiness_per.append(data["clouds"]["all"])
            wind_speeds.append(data["wind"]["speed"])
            weather_desc.append(data["weather"][0]["description"])

            
        except Exception as e:
            print(e)
            # print("this is happening")
            # append defaults to lists
            countries.append("")
            longs.append(np.nan)
            lats.append(np.nan)
            max_temps.append(np.nan)
            humidities.append(np.nan)
            cloudiness_per.append(np.nan)
            wind_speeds.append(np.nan)
            weather_desc.append(np.nan)
    else:
        print(response.status_code)
        print(response.text)
        countries.append("")
        longs.append(np.nan)
        lats.append(np.nan)
        max_temps.append(np.nan)
        humidities.append(np.nan)
        cloudiness_per.append(np.nan)
        wind_speeds.append(np.nan)
        weather_desc.append(np.nan)
        
    # sleep for one second
    time.sleep(1.1)
        
# build dataframe
# countries, longs, lats, max_temps, humidities, cloudiness_per, wind_speeds, weather_desc
df = pd.DataFrame()
df["City"] = cities
print(f"len of countries: {len(countries)}")
print(f"countries")
df["Country"] = countries
df["Lat"] = lats
df["Lng"] = longs
df["Max Temp"] = max_temps
df["Humidity"] = humidities
df["Cloudiness"] = cloudiness_per
df["Wind Speed"] = wind_speeds
df["Current Description"] = weather_desc


df
```

```python
df = df.dropna(axis=0)
df
```

```python
# Create the output file (CSV).
output_data_file = "WeatherPy_Database.csv"
# Export the City_Data into a CSV.
df.to_csv(output_data_file, index_label="City_ID")
```
